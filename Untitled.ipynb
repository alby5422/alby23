```python

from __future__ import print_function
import os
data_path = ['data']
```

```python
import pandas as pd

# Import the data using the file path
filepath = os.sep.join(data_path + ['predictive_maintenance.csv'])
data = pd.read_csv('predictive_maintenance.csv')
data

```

```python
data.drop(['Failure Type','Type'], axis=1, inplace=True)
data
```

```python
from sklearn.preprocessing import LabelBinarizer

lb = LabelBinarizer()

for col in ['Product ID']:
    data[col] = lb.fit_transform(data[col])

data
```

```python
import warnings
warnings.filterwarnings('ignore', module='sklearn')
```

```python
from sklearn.preprocessing import MinMaxScaler

msc = MinMaxScaler()

data = pd.DataFrame(msc.fit_transform(data),  
                    columns=data.columns)
data
```

```python
x_cols = [x for x in data.columns if x != 'Product ID']

# Split the data into two dataframes
X_data = data[x_cols]
y_data = data['Product ID']
X_data
y_data
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=3)

knn = knn.fit(X_data, y_data)

y_pred = knn.predict(X_data)
```

```python
def accuracy(real, predict):
    return sum(y_data == y_pred) / float(real.shape[0])





print(accuracy(y_data, y_pred))
```
